```python
import pandas as pd
import tiktoken
from glob import glob
# Supongamos que tienes un JSON guardado en un archivo llamado 'data.json'
dir_files = glob('batch_*.json')
# Lista de archivos

# Ordenar la lista por el n√∫mero en el nombre del archivo
sorted_files = sorted(dir_files, key=lambda x: int(x.split('_')[1].split('.')[0]))

print(sorted_files)
```

```python
import pandas as pd
import tiktoken
import json


# Define la funci√≥n para contar los tokens
def num_tokens_from_string(string: str, encoding_name: str) -> int:
    """Retorna el n√∫mero de tokens en un string de texto."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens

# Lista para almacenar los datos procesados


# Procesa cada archivo JSON
for idx, dir_file in enumerate(sorted_files):
    all_data = []
    print(f'Procesando el archivo: {idx + 1}/{len(sorted_files)} - {dir_file}')
    
    # Lee el archivo JSON en un DataFrame
    df = pd.read_json(dir_file)

    # Aplica la funci√≥n para contar los tokens en la columna 'content'
    df['num_tokens'] = df['content'].apply(lambda x: num_tokens_from_string(x, "cl100k_base"))

    # Filtra las filas con cero tokens
    df_filtered = df[df['num_tokens'] > 0]

    # Opcional: Elimina la columna 'num_tokens' si ya no la necesitas
    df_filtered = df_filtered.drop(columns=['num_tokens'])

    # Convierte el DataFrame filtrado a una lista de diccionarios
    data_list = df_filtered.to_dict(orient='records')

    # A√±ade la lista de diccionarios a la lista general
    all_data.extend(data_list)
    # Guarda la lista completa de diccionarios en un archivo JSON
    with open(f'data/batch_{idx}.json', 'w') as f:
        json.dump(all_data, f, indent=4)


```
